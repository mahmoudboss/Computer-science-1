```python
# monvies librery
# step 1: giving a name foe the section
movie_library = []

# step 2: Add Movies to the Library
movie_library.append({
    'title': 'Inception',
    'year': 2010,
    'genres': ['Action', 'Sci-Fi'],
    'rating': (4.5, 1500),
    'reviews': {'Alice': 'Mind-blowing!', 'Bob': 'Amazing concept.'}
})

movie_library.append({
    'title': 'Titanic',
    'year': 1997,
    'genres': ['Romance', 'Drama'],
    'rating': (4.7, 2000),
    'reviews': {'Eve': 'So emotional.', 'Tom': 'Classic love story.'}
})

movie_library.append({
    'title': 'The Matrix',
    'year': 1999,
    'genres': ['Action', 'Sci-Fi'],
    'rating': (4.8, 1800),
    'reviews': {'Sam': 'Red pill or blue pill?', 'Jill': 'Iconic scenes.'}
})


# Step 3: define the function : View All Movies
# This function loops will give us information for each movie.

def view_all_movies():
    for movie in movie_library:
        title = movie['title']
        year = movie['year']
        genres = ", ".join(movie['genres'])  # Convert genre list to a string by use .join to be more easier
        rating, num_reviews = movie['rating']
        # We use an f-string to neatly print the movie info
        print(f"Title: {title}, Year: {year}, Genres: {genres}, Rating: {rating}, Reviews: {num_reviews}")


# Step 4: define the function : Search Movies by Genre

def search_by_genre(genre):
    found = False  # Use for traking for movies if found will give us the resault if no will print not found
    for movie in movie_library:
        if genre in movie['genres']:  # Check if the genre is in the movie's genres list and when found
            print(f"- {movie['title']} ({movie['year']})")  # Show matching movie
            found = True
    if not found:
        print("No movies found in this genre.")  # Tell the user not founded


# Step 5: define the function : view movie details using the tilte
# this functin will give all the information about the movie when we write the title .

def view_movie_details(title):
    for movie in movie_library: # if the movie found in the library
        if movie['title'].lower() == title.lower():  # for that reason we use .lower "Inception" and "inception" will be treated as the same.
            print(f"Title: {movie['title']}")
            print(f"Year: {movie['year']}")
            print(f"Genres: {', '.join(movie['genres'])}")
            rating, num_reviews = movie['rating']
            print(f"Rating: {rating} ({num_reviews} reviews)")
            print("Reviews:")
            # Print each review in the reviews dictionary
            for reviewer, review in movie['reviews'].items(): # this give as information about the reviewer , what was the review and .item give us the information for each review
                print(f"- {reviewer}: {review}")
            return
    print("Movie not found.")  # Message if movie title is not found

#step 6 :
def add_review(title, reviewer_name, review_text):
    for movie in movie_library:
        # We use .lower() so that it doesn't matter if the user types capital letters or not.
        # This makes sure 'Inception' and 'inception' are treated the same.
        if movie['title'].lower() == title.lower():
            # This line adds the review to the movie.
            # The person's name is the key, and their comment is the value.
            # For example: reviews['Alice'] = "Great movie!"
            movie['reviews'][reviewer_name] = review_text

            # Now we update the rating.
            current_rating, current_reviews = movie['rating']  # Get the current average and how many reviews there are
            total_score = current_rating * current_reviews  # Multiply to get the total score so far
            total_score += 5  # Add 5 points for the new review (+= just means add to what was already there)
            current_reviews += 1  # Add one more to the total number of reviews
            new_rating = total_score / current_reviews  # Divide to get the new average rating

            # Save the new rating and review count back into the movie data
            movie['rating'] = (new_rating, current_reviews)

            # f"..." lets us put variables directly into a string.
            # {:.2f} means round the number to 2 decimal places, like 4.68
            print(f"Review added for '{movie['title']}' by {reviewer_name}.")
            print(f"New rating: {new_rating:.2f} stars based on {current_reviews} reviews.")
            return

    # If the movie title isn't found in the list, show this message
    print("Movie not found.")


#step 7 : This function calculates the average rating of all movies combined.

def calculate_average_rating():
    total_score = 0  # This keeps track of the combined score from all movies (starts at zero)
    total_reviews = 0  # This counts the total number of reviews from all movies (starts at zero)

    for movie in movie_library:
        rating, num_reviews = movie['rating']  # rating is a float, num_reviews is an integer
        total_score += rating * num_reviews  # += adds the result to the existing total_score
        # Example: total_score = total_score + (rating * num_reviews)
        total_reviews += num_reviews  # Same idea here: we add each movie's review count

    if total_reviews == 0:  # This checks if there are zero total reviews (== means 'is equal to')
        print("No reviews available to calculate average rating.")
    else:
        average_rating = total_score / total_reviews  # Calculates the overall average rating
        # {:.2f} means format the number to 2 decimal places. Example: 4.6666 becomes 4.67
        print(f"\nAverage rating of all movies: {average_rating:.2f} stars")



# step8 : View all movies
# This prints a summary (title, year, genres, rating, and number of reviews) for each movie in the library.
print("All Movies:")
view_all_movies()

# Search by genre
# '\n' adds a blank line before the output just to keep things looking clean in the terminal.
# This part will show all movies that include the genre 'Sci-Fi' in their genre list.
print("\nSearching for 'Sci-Fi' movies:")
search_by_genre('Sci-Fi')

# View details of a specific movie
# This shows everything about a movie: title, year, genres, rating info, and each individual review.
print("\nMovie Details for 'Inception':")
view_movie_details('Inception')

# Add a new review
# This adds a new review with a default 5-star score to the movie and recalculates the average rating.
print("\nAdding a review to 'Inception'...")
add_review('Inception', 'Charlie', 'A masterpiece of modern cinema.')

# Check details again to see the update
# This confirms that the new review was added by reprinting the full movie info.
print("\nUpdated Movie Details for 'Inception':")
view_movie_details('Inception')

# Calculate overall average rating
# This calculates the total average rating across all movies in the library based on all reviews.
print("\nCalculating overall average rating:")
calculate_average_rating()
```

```python
#pandas project
# Step 1: Import Pandas library
# (We need pandas to work with tables and data)
import pandas as pd

# Step 2: Create a DataFrame with sample data
# (A DataFrame is like a table made of rows and columns)
data = {
    'Name': ['John', 'Alice', 'Bob', 'Charlie', 'David'],
    'Age': [23, 30, 25, 35, 28],
    'City': ['New York', 'Los Angeles', 'Chicago', 'San Francisco', 'New York'],
    'Salary': [50000, 70000, 45000, 80000, 55000]
}

# Creating the DataFrame using the data dictionary
df = pd.DataFrame(data)

# Step 3: Print the DataFrame to see the full table
print("Full DataFrame:")
print(df)
print("\n")  # Newline for better readability

# Step 4: Basic Operations

# a. Display only the "Name" and "Salary" columns
print("Name and Salary columns:")
print(df[['Name', 'Salary']])
print("\n")

# b. Show the first 3 rows of the DataFrame
print("First 3 rows of the DataFrame:")
print(df.head(3))  # head(3) shows the first 3 rows
print("\n")

# c. Find the average salary
average_salary = df['Salary'].mean()
print("Average Salary:", average_salary)
print("\n")

# d. Filter and show rows where Age is greater than or equal to 30
print("People aged 30 or older:")
print(df[df['Age'] >= 30])
print("\n")

# Step 5: Sorting

# Sort by "Salary" in descending order
print("Sorted by Salary (highest first):")
print(df.sort_values(by='Salary', ascending=False))
print("\n")

# Sort by "City" and then "Age" in ascending order
print("Sorted by City and then by Age:")
print(df.sort_values(by=['City', 'Age'], ascending=True))
print("\n")

# Step 6: Working with Missing Data

# a. Add missing values manually
df.loc[2, 'Salary'] = None   # Set Bob's Salary to None (missing)
df.loc[4, 'City'] = None     # Set David's City to None (missing)

# Display DataFrame after introducing missing values
print("DataFrame with Missing Values:")
print(df)
print("\n")

# b. Fill missing Salary values with the average salary
# First, calculate the current average salary (ignoring missing values)
average_salary = df['Salary'].mean()
# Now fill missing salaries with this average
df['Salary'].fillna(average_salary, inplace=True)

# c. Fill missing City values with "Unknown"
df['City'].fillna('Unknown', inplace=True)

# Display DataFrame after filling missing values
print("DataFrame after filling missing values:")
print(df)
print("\n")

# d. Drop rows with any missing values (this will not drop anything now because we filled missing data)
# (If we didn't fill, this would remove rows that had None)
df_cleaned = df.dropna()

print("DataFrame after dropping missing values:")
print(df_cleaned)
print("\n")

# Step 7: Save the cleaned data into a CSV file
df_cleaned.to_csv('Cleaned_data.csv', index=False)  # Save without row numbers

print("Data has been saved to Cleaned_data.csv!")
print("\n")

# Bonus Task: Group By City and find the average salary
print("Average Salary Grouped by City:")
print(df_cleaned.groupby('City')['Salary'].mean())
```

```python
# Project: Personal Budget Tracker

# Create empty lists to store income and expense entries
incomes = []
expenses = []

# Step 2: Define the function to add income
def add_income():
    print("\nAdd Income")
    description = input("Enter income description: ")
    amount = float(input("Enter income amount: "))
    income_entry = {"description": description, "amount": amount}
    incomes.append(income_entry)
    print("Income added successfully!")

# Step 3: Define the function to add an expense
def add_expense():
    print("\nAdd Expense")
    description = input("Enter expense description: ")
    amount = float(input("Enter expense amount: "))
    category = input("Enter expense category: ")
    expense_entry = {"description": description, "amount": amount, "category": category}
    expenses.append(expense_entry)
    print("Expense added successfully!")

# Step 4: Define the function to view a budget summary
def view_summary():
    print("\nBudget Summary")
    total_income = sum(income["amount"] for income in incomes)
    total_expense = sum(expense["amount"] for expense in expenses)
    balance = total_income - total_expense
    print("Total Income: $", total_income)
    print("Total Expenses: $", total_expense)
    print("Balance: $", balance)

# Step 5: Define the function to delete an entry
def delete_entry():
    print("\nDelete Entry")
    choice = input("Delete income or expense? (income/expense): ")
    if choice == "income":
        description = input("Enter the income description to delete: ")
        for income in incomes:
            if income["description"].lower() == description.lower():
                incomes.remove(income)
                print("Income deleted successfully!")
                return
        print("Income not found.")
    elif choice == "expense":
        description = input("Enter the expense description to delete: ")
        for expense in expenses:
            if expense["description"].lower() == description.lower():
                expenses.remove(expense)
                print("Expense deleted successfully!")
                return
        print("Expense not found.")
    else:
        print("Invalid choice.")

# Step 6: Define the function to modify an entry
def modify_entry():
    print("\nModify Entry")
    choice = input("Modify income or expense? (income/expense): ")
    if choice == "income":
        description = input("Enter the income description to modify: ")
        for income in incomes:
            if income["description"].lower() == description.lower():
                new_description = input("Enter new description: ")
                new_amount = float(input("Enter new amount: "))
                income["description"] = new_description
                income["amount"] = new_amount
                print("Income modified successfully!")
                return
        print("Income not found.")
    elif choice == "expense":
        description = input("Enter the expense description to modify: ")
        for expense in expenses:
            if expense["description"].lower() == description.lower():
                new_description = input("Enter new description: ")
                new_amount = float(input("Enter new amount: "))
                new_category = input("Enter new category: ")
                expense["description"] = new_description
                expense["amount"] = new_amount
                expense["category"] = new_category
                print("Expense modified successfully!")
                return
        print("Expense not found.")
    else:
        print("Invalid choice.")

# Step 7: Define the function to list all income and expense entries
def list_entries():
    print("\nList of Entries")
    if incomes:
        print("\nIncomes:")
        for income in incomes:
            print("Description:", income["description"], "| Amount: $", income["amount"])
    else:
        print("\nNo income entries.")

    if expenses:
        print("\nExpenses:")
        for expense in expenses:
            print("Description:", expense["description"], "| Amount: $", expense["amount"], "| Category:", expense["category"])
    else:
        print("\nNo expense entries.")

# Step 8: Create the main function
def main():
    while True:
        print("\nPersonal Budget Tracker")
        print("1. Add Income")
        print("2. Add Expense")
        print("3. View Budget Summary")
        print("4. Delete an Entry")
        print("5. Modify an Entry")
        print("6. List All Entries")
        print("7. Exit")

        choice = input("Enter your choice (1-7): ")

        if choice == "1":
            add_income()
        elif choice == "2":
            add_expense()
        elif choice == "3":
            view_summary()
        elif choice == "4":
            delete_entry()
        elif choice == "5":
            modify_entry()
        elif choice == "6":
            list_entries()
        elif choice == "7":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

# Step 9: Run the main function
main()

```

```python
#restaurent project
# Dish class represents a dish on the menu
class Dish:
    def __init__(self, name, price, category):
        # 'self' refers to the current object
        self.name = name
        self.price = price
        self.category = category

    def __str__(self):
        # This is used when you print the Dish object
        # It returns the dish in a nice readable format
        return f"{self.name} ({self.category}) - ${self.price:.2f}"


# Order class represents a customer's order
class Order:
    def __init__(self, order_id, customer):
        self.order_id = order_id               # Unique ID for each order
        self.customer = customer               # The customer who placed the order
        self.ordered_dishes = []               # List of Dish objects
        self.status = "Pending"                # Default status for new orders

    def add_dish(self, dish):
        # Adds a selected dish to the order list
        self.ordered_dishes.append(dish)

    def calculate_total(self):
        # Adds up the prices of all dishes in the order
        total = sum(dish.price for dish in self.ordered_dishes)
        return total

    def view_order(self):
        # Prints order details in a readable way
        print(f"\nOrder ID: {self.order_id}")  # '\n' adds a blank line for clarity
        print(f"Customer: {self.customer.name} | Email: {self.customer.email}")
        print("Ordered Dishes:")
        for dish in self.ordered_dishes:
            print(f" - {dish}")  # Uses __str__ of the Dish class
        print(f"Total Price: ${self.calculate_total():.2f}")
        print(f"Status: {self.status}")


# Customer class represents a person placing an order
class Customer:
    def __init__(self, name, email):
        self.name = name
        self.email = email
        self.order_history = []  # Keeps track of all orders by this customer

    def place_order(self, restaurant, dish_list):
        # Create a new order with a unique ID
        # The ID is the current number of orders + 1 (to avoid ID = 0)
        order_id = len(restaurant.orders) + 1  # <-- This is the +1 you asked about
        order = Order(order_id, self)

        for dish in dish_list:
            order.add_dish(dish)

        self.order_history.append(order)       # Save order in customer's history
        restaurant.place_order(order)          # Add the order to the restaurant's records
        print(f"\nOrder placed successfully! (Order ID: {order_id})")
        return order


# Restaurant class holds menu and manages orders
class Restaurant:
    def __init__(self):
        self.menu = []    # List of all available Dish objects
        self.orders = []  # List of all Order objects

    def add_dish_to_menu(self, dish):
        # Add a dish to the menu
        self.menu.append(dish)

    def view_menu(self):
        print("\n--- MENU ---")  # '\n' adds a space before the menu title
        for dish in self.menu:
            print(dish)  # Uses Dish class's __str__

    def place_order(self, order):
        # Add a new order to the restaurant's order list
        self.orders.append(order)

    def view_orders(self):
        print("\n--- ALL ORDERS ---")  # Neatly formats the start of the order list
        for order in self.orders:
            order.view_order()


# ------- Example Usage --------

# Create a restaurant
my_restaurant = Restaurant()

# Add dishes to the menu
dish1 = Dish("pasta", 12.99, "Main Course")
dish2 = Dish("algerian Salad", 8.50, "Starter")
dish3 = Dish("Chocolate Cake", 5.75, "Dessert")
dish4 = Dish("Lemonade", 2.99, "Drink")

my_restaurant.add_dish_to_menu(dish1)
my_restaurant.add_dish_to_menu(dish2)
my_restaurant.add_dish_to_menu(dish3)
my_restaurant.add_dish_to_menu(dish4)

# View the menu
my_restaurant.view_menu()

# Create a customer
customer1 = Customer("mahmoud", "mahmoud@example.com")

# Customer chooses dishes to order
selected_dishes = [dish1, dish3, dish4]  # Chosen from the menu

# Place the order and view it
order1 = customer1.place_order(my_restaurant, selected_dishes)
order1.view_order()

# Restaurant staff views all orders
my_restaurant.view_orders()
```

```python
# Simple Banking System in Python

# ---- User Info (just a basic setup with username and password) ----
username = "user123"
password = "pass123"
balance = 1000.0  # This is how much money the user has at the beginning

# ---- Login Part ----
attempts = 0  # Counts how many times user tried to log in

# We let the user try 3 times to log in
while attempts < 3:
    input_username = input("Enter username: ")
    input_password = input("Enter password: ")

    # Check if what they typed matches the correct login
    if input_username == username and input_password == password:
        print("\nLogin successful!\n")
        break  # Stop asking for login if correct
    else:
        print("Incorrect username or password. Try again.")
        attempts += 1  # Add 1 to attempts if login is wrong

# If they fail 3 times, stop the program
if attempts == 3:
    print("\nToo many failed attempts. Exiting system.")
else:
    # ---- Main Menu (user sees this after logging in) ----
    while True:
        print("---- Main Menu ----")
        print("1. Check Balance")
        print("2. Deposit Money")
        print("3. Withdraw Money")
        print("4. Exit")

        choice = input("Choose an option (1-4): ")

        # ---- Option 1: See how much money you have ----
        if choice == "1":
            print(f"\nYour current balance is: ${balance:.2f}\n")

        # ---- Option 2: Add money to your account ----
        elif choice == "2":
            try:
                amount = float(input("Enter amount to deposit: "))
                if amount > 0:
                    balance += amount  # Add the amount to current balance
                    print(f"${amount:.2f} deposited successfully!")
                    print(f"New balance: ${balance:.2f}\n")
                else:
                    print("Deposit amount must be positive.\n")
            except ValueError:
                print("Invalid input. Please enter a number.\n")

        # ---- Option 3: Take money out ----
        elif choice == "3":
            try:
                amount = float(input("Enter amount to withdraw: "))
                if amount <= 0:
                    print("Withdrawal amount must be positive.\n")
                elif amount > balance:
                    print("Insufficient funds!\n")  # Can't take more than you have
                else:
                    balance -= amount  # Take out the money
                    print(f"${amount:.2f} withdrawn successfully!")
                    print(f"New balance: ${balance:.2f}\n")
            except ValueError:
                print("Invalid input. Please enter a number.\n")

        # ---- Option 4: Leave the system ----
        elif choice == "4":
            print("Thank you for using the banking system. Goodbye!")
            break  # Exit the loop and stop the program

        # ---- If the user types something wrong ----
        else:
            print("Invalid choice. Please enter a number between 1 and 4.\n")
```
