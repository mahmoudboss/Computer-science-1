```python
# This is a simple program that helps you decide what to wear based on the weather.
# It also asks you to log in before using it.

# Step 1: Set up a correct username and password that the program will accept.
correct_username = "mahmoud"
correct_password = "boss"

# This loop lets the user try to log in up to 3 times
# range(3) means it will run 3 times: 0, 1, and 2
for attempt in range(3):
    # Ask the user to enter their username and password
    username = input("Enter username: ")
    password = input("Enter password: ")

    # Check if both the username and password are correct
    if username == correct_username and password == correct_password:
        print("Login successful! Welcome!\n")
        break  # break means we stop the loop early if the login is successful
    else:
        print("Incorrect username or password. Try again.\n")
else:
    # This runs only if the loop finishes without a successful login
    print("Too many failed attempts. Exiting the program.")
    exit()  # exit() ends the program completely

# Step 2: Set up empty variables to store weather information
# = None means the variable currently has nothing in it – like an empty box waiting for input
temperature = None
is_raining = None
is_windy = None

# Step 3: Keep showing the menu until the user wants to exit
while True:
    # Show the main menu
    print("\nMain Menu:")
    print("1. Enter today’s weather")
    print("2. Get clothing suggestion")
    print("3. Exit")

    # Ask the user to choose a menu option
    choice = input("Choose an option (1, 2, or 3): ")

    # If the user chooses option 1, ask for the weather
    if choice == "1":
        try:
            # try: means "try to do something that might cause an error"
            # float(input(...)) asks the user to type something and turns it into a number
            # float means the number can have a decimal point, like 15.5
            temperature = float(input("Enter the temperature in Celsius: "))
        except ValueError:
            # except is used to catch problems that happen in the try part
            # ValueError happens if the user types something that isn't a number (like "hot")
            print("Please enter a valid number for temperature.")
            continue  # continue means skip the rest of the code and go back to the menu

        # Ask if it's raining or windy
        # .strip() removes any spaces before or after the word
        # .lower() turns the text into all lowercase letters (e.g., "YES" becomes "yes")
        is_raining = input("Is it raining? (yes or no): ").strip().lower()
        is_windy = input("Is it windy? (yes or no): ").strip().lower()

        # Check if the answers are valid – only 'yes' or 'no' are allowed
        if is_raining not in ["yes", "no"] or is_windy not in ["yes", "no"]:
            print("Please answer only with 'yes' or 'no'.")
            # Reset values back to None because they were invalid
            temperature = None
            continue  # go back to the menu

        print("Weather information saved.")

    # If the user chooses option 2, give clothing advice
    elif choice == "2":
        # choice == "2" means we're checking if the user typed "2"

        # If they haven't entered the weather yet, remind them to do it
        if temperature is None:
            print("Please enter the weather first (Option 1).")
            continue  # go back to the menu

        # Suggest clothes based on the temperature
        if temperature < 10:
            print("It's cold. Wear a coat.")
        elif 10 <= temperature < 20:
            print("It's mild. A light jacket should be fine.")
        else:
            print("It's warm. A t-shirt is enough.")

        # Add extra suggestions if it's raining or windy
        if is_raining == "yes":
            print("Don't forget your umbrella!")
        if is_windy == "yes":
            print("Wear something that blocks the wind.")

    # If the user chooses option 3, exit the program
    elif choice == "3":
        print("Goodbye!")
        break  # break stops the while loop and ends the program

    # If they type something that's not 1, 2, or 3
    else:
        print("Invalid choice. Please select 1, 2, or 3.")


```
